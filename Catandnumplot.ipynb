```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

This is the dataframe: https://www.kaggle.com/datasets/madhurpant/top-650-fifa-22-players-simplified

```python
df = pd.read_csv('Fifa_players.csv')
```

```python
df.head()
```

```python
fig, ax = plt.subplots(figsize=(5,6))

sns.boxplot(x=df.Position, y=df.attacking, ax=ax)
```

```python
fig, ax = plt.subplots(figsize=(5,6))

sns.violinplot(x=df.Position, y=df.attacking, ax=ax)
```

```python
fig, ax = plt.subplots(figsize=(5,6))

sns.stripplot(x=df.Position, y=df.attacking, jitter=True, ax=ax, alpha=0.5)
```

```python
fig, ax = plt.subplots(figsize=(5,6))

sns.swarmplot(x=df.Position, y=df.attacking, ax=ax, size=4)
```
