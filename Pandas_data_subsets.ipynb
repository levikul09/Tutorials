```python
# Or use dictionary to get series

df['long_name']
```

```python
df['power'].mean()
```

```python
#Pass list to select multiple columns

df[['long_name', 'short_name', 'power']]
```

```python
# Use the .get method to 

df.get('power', False)
```

```python
df.get('power', False).mean()
```

```python
df.get('asd', "No such column")
```

```python
#use string methods

df[[col for col in df.columns if col.endswith('foot')]] #select all columns that ends with foot
```

```python
#use string methods

df[['short_name'] + [col for col in df.columns if col.endswith('foot')]] #Select name + all columns that ends with foot 
```

```python
df[
    ['short_name'] +                                         #Select name
   [col for col in df.columns if col.endswith('foot')] +    #Select all columns that ends with foot
   [col for col in df.columns if col.endswith('ing')]       #Select all columns that ends with ing
   ]      
```

```python
# Get every column wth the exception of short name and long name

df[
    df.columns.difference(['short_name', 'long_name'])
]
```

```python
# Use .isin to subset columns based on list

df.columns.isin(['power', 'skill'])
```

```python
df[df.columns[df.columns.isin(['power', 'skill'])]]
```

```python
df[df.columns[~df.columns.isin(['nationality_name', 'club_jersey_number',
       'preferred_foot', 'weak_foot', 'skill_moves', 'player_traits', 'pace',
       'shooting', 'passing', 'dribbling', 'defending', 'physic', 'attacking',
       'skill', 'movement', 'power', 'mentality'])]]

# Use the tilde as a negation operator to reverse the selection. It inverses True to False and False to True
```

```python
df[df.columns[df.dtypes != 'object']] # Select columns based on dtypes that are not objects
```

```python
df[df.columns[df.dtypes == 'int64']] # Select columns based on dtypes that are integers
```
