```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
chance=1/7.5 + 1/15 + 1/30 + 1/20 + 1/300 + 1/40800 + 1/408000 + 1/2550000

chance
```

```python
print(f'The Expected payoff is: ${1/7.5*150 + 1/15*200 + 1/30*300 + 1/20*500 + 1/300*1000 + 1/40800*10000 + 1/408000*100000 + 1/2550000 *20000000}')
```

```python
avg_payoff =[]

n=1

while n<1000:
    win=random.choices([0,150, 200, 300, 500, 1000, 10000, 100000, 20000000], weights=[1-chance,1/7.5,1/15,1/30,1/20,1/300,1/40800,1/408000,1/2550000], k=n)
    avg_payoff.append(np.mean(win))
    n+=1
    
x=range(1,1000)
    
plt.plot(x, avg_payoff)
plt.hlines(y=80, xmin=0, xmax=1000, color='r')
```

```python
avg_dice =[]

n=1

while n<1000:
    throws=random.choices([1,2,3,4,5,6], weights=[1/6,1/6,1/6,1/6,1/6,1/6], k=n)
    avg_dice.append(sum(throws)/n)
    n+=1
    
x=range(1,1000)
    
plt.plot(x, avg_dice)
plt.hlines(y=3.5, xmin=0, xmax=1000, color='r')
```
