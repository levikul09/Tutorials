```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
```

The lottery ticket I used: https://www.texaslottery.com/export/sites/lottery/Games/Scratch_Offs/details.html_252701659.html

```python
chance=1/8.5 + 1/16 + 1/31 + 1/21 + 1/301 + 1/40801 + 1/408001 + 1/2550001

chance
```

```python
expected_payoff = (1/8.5*150 + 1/16*200 + 1/31*300 + 1/21*500 + 1/301*1000 +
                    1/40801*10000 + 1/408001*100000 + 1/2550001 *20000000)

print(f'The Expected payoff is: ${expected_payoff}')
```

```python
fig, ax=plt.subplots(figsize=(10,5))

avg_payoff =[]

n=1

while n<1000:
    win=random.choices([0,150, 200, 300, 500, 1000, 10000, 100000, 20000000], weights=[1-chance,1/8.5,1/16,1/31,1/21,1/301,1/40801,1/408001,1/2550001], k=n)
    avg_payoff.append(np.mean(win))
    n+=1
    
x=range(1,1000)

plt.plot(x, avg_payoff)
plt.hlines(y=expected_payoff, xmin=0, xmax=1000, color='r');
```

# LLN with Dice Rolls 

```python
fig, ax=plt.subplots(figsize=(8,5))

avg_dice =[]

n=1

while n<10000:
    throws=random.choices([1,2,3,4,5,6], weights=[1/6,1/6,1/6,1/6,1/6,1/6], k=n)
    avg_dice.append(sum(throws)/n)
    n+=1
    
x=range(1,10000)
    
plt.plot(x, avg_dice)
plt.hlines(y=3.5, xmin=0, xmax=10000, color='r')

plt.title('Law of large numbers')

#plt.savefig('dicerolls', dpi=300, facecolor='white',bbox_inches='tight' )
```
