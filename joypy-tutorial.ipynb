```python
import numpy as np
import pandas as pd
import joypy
import matplotlib.pyplot as plt
from matplotlib import cm
import datetime as dt
```

2. Modify data 

```python
df = pd.read_csv('/kaggle/input/szeged-weather/weatherHistory.csv')
df['Formatted Date'] = df['Formatted Date'].str[:10]
df['Formatted Date'] = pd.to_datetime(df['Formatted Date'], format="%Y-%m-%d")
year = df['Formatted Date'].dt.year
```

3. Plot ðŸ“ˆ

```python
fig, ax = joypy.joyplot(df, by=year, column="Temperature (C)", overlap=0.8, 
                        title="Weather in Szeged 2006-2016")
```
