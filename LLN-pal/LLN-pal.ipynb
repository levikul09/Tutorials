```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import random
```

# Dice


```python
exp = (1+2+3+4+5+6)/6
```

```python
exp
```

```python
avg_dice =[]

n=1

while n<=10000:
    np.random.seed(11)
    throws=np.random.randint(1,7,size=n)
    avg_dice.append(sum(throws)/n)
    n+=1
    
avg_dice
```

```python
fig, ax = plt.subplots(figsize=(10,5), facecolor='#1E1E1E')

#Plotting:

y = avg_dice[:10000]

ax.plot(y, color='#D81E5B', linewidth=3)

ax.hlines(exp, 0, len(y), color='#E8C547')

#add text to right upper corner

ax.text(len(y)/2*0.6, 5.3, f'No. of experiments: {len(y)}', color='#06A062', fontsize=16, fontweight='bold')
ax.text(len(y)/2*0.6, 4.8, f'Mean of means: {round(np.mean(y), 2)}', color='#06A062', fontsize=16, fontweight='bold')

if len(y)<=10:
    ax.text(0, exp+0.2, f'Expected Value: {exp}', color='#E8C547', fontsize=16, fontweight='bold')

#Formatting: 

ax.tick_params(which='both', colors='white')

ax.spines[['left', 'bottom']].set_color('white')

ax.spines[['right', 'top']].set_visible(False)

ax.set_facecolor('#1E1E1E')

ax.set_yticks([1,2,3,4,5,6])

if len(y)<=10:
    ax.set_xticks(np.arange(len(y)), np.arange(1, len(y)+1))

plt.savefig(f'{len(y)}runs.png', dpi=300,bbox_inches='tight');
```

# Lottery

```python
chance=1/8.5 + 1/16 + 1/31 + 1/21 + 1/301 + 1/40801 + 1/408001 + 1/2550001

print(f'The chance to win something is: {chance:.2%}')
```

```python
expected_payoff = (1/8.5*150 + 1/16*200 + 1/31*300 + 1/21*500 + 1/301*1000 +
                    1/40801*10000 + 1/408001*100000 + 1/2550001 *20000000)

print(f'The Expected payoff is: ${expected_payoff}')
```

```python
np.random.seed(10)

num_iterations = 100000000

# Define the payoff options and their respective probabilities
payoff_options = np.array([0, 150, 200, 300, 500, 1000, 10000, 100000, 20000000])
probabilities = np.array([1 - chance, 1 / 8.5, 1 / 16, 1 / 31, 1 / 21, 1 / 301, 1 / 40801, 1 / 408001, 1 / 2550001])

# Simulate all iterations at once
win = np.random.choice(payoff_options, size=(num_iterations,), p=probabilities)

# Calculate the average payoff for each iteration
avg_payoff = np.cumsum(win) / np.arange(1, num_iterations + 1)
```

```python
fig, ax = plt.subplots(figsize=(10,5), facecolor='black')

#Plotting:

y3 = avg_payoff[:10000000]

ax.plot(y3, color='#D81E5B', linewidth=3)

ax.hlines(expected_payoff, 0, len(y3), color='#E7C547')

#add text to right upper corner

ax.text(len(y3)/2*0.6, 140, f'No. of experiments: {len(y3)}', color='#04A061', fontsize=16, fontweight='bold')
ax.text(len(y3)/2*0.6, 120, f'Mean of means: {round(np.mean(y3), 2)}', color='#04A061', fontsize=16, fontweight='bold')

if len(y3)<=10:
    ax.text(0, expected_payoff+0.2, f'expected Value: {round(expected_payoff, 2)}', color='#E7C547', fontsize=16, fontweight='bold')

#Formatting: 

ax.tick_params(which='both', colors='white')

ax.spines[['left', 'bottom']].set_color('white')

ax.spines[['right', 'top']].set_visible(False)

ax.set_facecolor('black')

#ax.set_yticks([20,40,60,75,80,120])

if len(y3)<=10:
    ax.set_xticks(np.arange(len(y3)), np.arange(1, len(y3)+1))

#plt.savefig(f'{len(y3)}runs.png', dpi=300,bbox_inches='tight');
```
