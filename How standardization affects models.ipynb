```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Creating a synthetic dataset
np.random.seed(42)  # For reproducibility
weights = np.random.uniform(50, 150, 100)  # 100 individuals with weights between 50kg and 150kg
heights = np.random.uniform(1.5, 2.0, 100)  # Heights between 1.5m and 2.0m

# Combining the features into a dataframe
df = pd.DataFrame({
    'Weight': weights,
    'Height': heights
})

# Applying K-Means clustering on the original dataset
kmeans_orig = KMeans(n_clusters=4, random_state=42).fit(df)

# Standardizing the data
scaler = StandardScaler()
df_standardized = scaler.fit_transform(df)

# Applying K-Means clustering on the standardized dataset
kmeans_std = KMeans(n_clusters=4, random_state=42).fit(df_standardized)

# Visualizing the results
# Original Data
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
plt.scatter(df['Weight'], df['Height'], c=kmeans_orig.labels_, cmap='viridis')
plt.title('Original Data Clustering')
plt.xlabel('Weight (kg)')
plt.ylabel('Height (m)')

# Standardized Data
plt.subplot(1, 2, 2)
plt.scatter(df_standardized[:, 0], df_standardized[:, 1], c=kmeans_std.labels_, cmap='viridis')
plt.title('Standardized Data Clustering')
plt.xlabel('Weight (standardized)')
plt.ylabel('Height (standardized)')

plt.tight_layout()
plt.show()
```
